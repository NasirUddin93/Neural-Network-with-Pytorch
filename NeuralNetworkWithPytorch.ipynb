```python
class CustomNeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        #layer
        self.conv1 = nn.Conv2d(in_channels = 1, out_channels = 6,kernel_size=5)
        self.conv1 = nn.Conv2d(in_channels = 6, out_channels = 12,kernel_size=5)
        
        #Fully connedted, Neural Network
        self.fcn1 = nn.Linear(in_features =12*4*4, out_features =120 )
        self.fcn2 = nn.Linear(in_features =12*4*4, out_features =60 )
        
         # Flatten() --> Connected Layer or Bridge Layer 2D - 1D
        self.outNode = nn.Linear(in_features = 60,out_features = 10)
        
        def forwardPropagation(self,ti):
            return ti
```

```python
network = CustomNeuralNetwork() # assign 
```

```python
class CustomNeuralNetwork2(nn.Module):
    def __init__(self):
        super().__init__()
        #layer
        self.conv1 = nn.Conv2d(in_channels = 1, out_channels = 6,kernel_size=5)
        self.conv1 = nn.Conv2d(in_channels = 6, out_channels = 12,kernel_size=5)
        
        #Fully connedted, Neural Network
        self.fcn1 = nn.Linear(in_features =12*4*4, out_features =120,bias=False )
        self.fcn2 = nn.Linear(in_features =12*4*4, out_features =60,bias = False )
        
         # Flatten() --> Connected Layer or Bridge Layer 2D - 1D
        self.outNode = nn.Linear(in_features = 60,out_features = 10,bias = False)
        
        def forwardPropagation(self,ti):
            return ti
```

```python
network2 = CustomNeuralNetwork2()
```
